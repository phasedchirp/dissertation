# Modeling on-line category learning with a locally Bayesian model

I attempt to model learning of phonetic categories using an itterative/online learning model which, while not a globally optimal/rational learner, implements Bayesian inference at the local level (within sub-components of the model).

The model is designed to satisfy several requirements:

* Online learning: The model takes in one stimulus item at a time, updating beliefs itteratively
* Preserves learner uncertainty at the local level: The learner does not learn the full joint distribution of parameters and data as in a globally Bayesian model, but does preserve uncertainty within sub-components of the model rather than making point estimates of parameters as in a maximum likelihood learner.
* Learn a category system without knowing the number of categories in advance.
* Adapt to a target distribution that potentially changes over time (non-stationary).

Additional goals in modeling category phonetic category acquisition in particular:

* Reasonably approximate the learning behavior observed in native-language category acquisition.
* Replicate a particular type of learner mistakes expected in learning of novel categories

### Modeling framework:

The model is built as an itteratively updated mixture of gaussians (MoG). Each category is represented as a component of the mixture with parameters $\mu$ (category mean), $\Sigma$ (covariance matrix), and $\phi$ (frequency/prior probability). The probability of a stimulus item $x$ for a mixture with $k$ categories is then
$$ p(x) = \sum_{i=1}^{k}\phi_i*\mathcal{N}(x|\mu_i,\Sigma_i)$$
while the probability that an item $x$ came from a category $c_i$ where $c_i\in \{c_1...c_k\}$ is
$$p(c_i|x) = \frac{p(x|c_i)}{\sum_{j=1}^{k}\phi_j*p(x|c_j)}$$

While the assumption that categories are normally distributed is too strong, it represents a convenient simplification rather than a necessary property of the model. The component distributions could easily be replaced by other parametric distributions or even non-parametric density estimates with loss only of computational convenience.

# Building the model:

The learner's category system is modeled as a mixture of normal distributions which have two parameters -- mean ($\mu$) and covariance matrix ($\Sigma$). To represent the learner's uncertainty around the parameters as it encounters new information, I'm using a normal inverse Wishart prior for the mean and covariance of each component ($p(\mu,\Sigma)\sim\mathcal{NW}^{-1}_{\kappa_0,\nu_0}(\mu_0,\Sigma_0)$. This is the conjugate prior distribution for a normal distribution with unknown mean and covariance, meaning that the distributions $p(\mu,\Sigma|data)$ have the same form as the prior $p(\mu,\Sigma)$ with the degree of freedom parameters $\kappa$ and $\nu$ of the prior being interpretable as degrees of confidence in the learner's prior beliefs about the mean and variance of each component.

First, defining data types for the mixture components and the mixture.

Components:
* N-dimensional gaussian distributions
  - Object is just storing parameters
  - Has the following fields:
    - d (dimension)
    - mu (mean)
    - sigma (covariance matrix)
    - kappa (degrees of confidence for the mean)
    - nu (degrees of confidence for the covariance matrix)
 
 Mixture:
 * Object storing an array of mixture components and mixing weights
 * Has the following fields:
   - d (dimension)
   - a (dispersion hyperparameter)
   - stick (current stick length for stick-breaking process)
   - alpha (parameter vector $\alpha$ for a Dirichlet prior on the mixing weights)
     - phi (mixing weights/parameter vector for multinomial, equal to $\mathrm{E}[Dirichlet(\alpha)]$)
   - components (array of components and mixing weights)
   - mean_sigma (scale matrix for hierarchical prior on component covariances)
   - ms_nu (degrees of freedom for mean_sigma)

```julia
type GaussND
    d::Int
    mu
    sigma
    kappa
    nu
end

type ItterMix
    d::Int
    a
    stick
    alpha
    phi
    components
    mean_sigma
    ms_nu
end
```

Next, defining update rules for the component distributions and the mixture as a whole.

* Update rules for components based on Gelman et. al 2013 section 3.6. Adapted for case where x is a single point.
  - covariance matrix $\Sigma$ has an Inverse Wishart ($W^{-1}$) distribution
  - mean $\mu$ has a normal distribution
* Update rules for mixing weights use the description of the Dirichlet distribution as a conjugate prior from Gelman et al. section 3.4 (p. 69).

Component update function is as follows (ordering of updates necessary):

* update scale matrix for $W^{-1}$ covariance prior:     $S_{new} = S_{old} + \frac{\kappa_{old}}{\kappa_{old}+1}(x-\mu_{old})(x-\mu_{old})^T$
* update mean of normal prior for the mean:     $\mu_{new} = \frac{\kappa_{old}}{\kappa_{old}+1}\mu_{old}+x*\frac{1}{\kappa_{old}+1}$
* update $\kappa$:     $\kappa_{new}=\kappa_{old}+1$
* update $\nu$:     $\nu_{new} = \nu_{old}+1$

Note that $\mathrm{E}[\Sigma]=\mathrm{E}[W_{\nu}^{-1}(S)]=\frac{S}{\nu-d-1}$ where $S$ is a $d$x$d$ matrix.

```julia
# component updates
function update(G::GaussND,x,weight = 1)
    G.sigma += ((x-G.mu)*transpose(x-G.mu))*G.kappa/(G.kappa+1);
    G.mu = G.mu*(G.kappa/(G.kappa+1)) + x*(1/(G.kappa+1));
    G.kappa += 1;
    G.nu += 1;
end
```

## Updating the mixture

Updating the mixture as a whole requires several steps:

* Determine whether an observation comes from a new or existing category
* If new:
  - Add a new component to the mixture
  - Determine an appropriate initial mixing weights for the newly observed component
* If old:
  - Assign the observation to an existing category (can do a winner-takes-all update or probabilistically weighted)
  - Adjust parameters of relevant component(s)
* Adjust mixing weights and prior over mixing weights

# Determining whether an observation comes from a new category:

Determining whether an observation comes from a new or existing category. Prior probabilities for the components in a finite mixture distribution have a (marginal) multinomial distribution. So, for a potentially infinite mixture model, I'm effectively treating unobserved categories as an additional possible outcome in addition to the previously observed categories. So, for example, in a learner that has observed 3 categories, the categories have the distribution:

$$p(C = i) \sim multinomial(\theta)$$

where $\theta$ is a vector

$$\begin{bmatrix}
\theta_1
\\
\theta_2
\\
\theta_3
\\
\theta_4
\end{bmatrix}$$

with $$\sum_{i=1}^4\theta_i=1$$

where the outcome corresponding to $\theta_4$ is observing a previously unobserved category. The observed categories then have the conditional distribution

$$p(C = i|\text{observed}) \sim multinomial\left(\begin{bmatrix}\frac{\theta_1}{\sum_1^3\theta_i}\\\frac{\theta_2}{\sum_1^3\theta_i}\\\frac{\theta_3}{\sum_1^3\theta_i}\end{bmatrix}\right)$$

where $\theta$ has been appropriately renormalized.

This is where what I'm doing gets a bit shaky. I'm basing this next part (loosely) on the stick-breaking construction for a Dirichlet process to give a reasonable prior over category weights. The stick breaking process works as follows -- Starting with a stick length of 1 and a dispersion parameter $alpha$:

* draw a random value $\beta_1 \sim beta(1,\alpha)$. This has an expected value $\frac{1}{1+\alpha}$
* break the stick at point $\beta_1$, giving a value $w_1$ which is the portion of the stick removed. (Remainder of the stick would be $\theta_4$ in the example above)
* repeat, calculating $w_i$ as $beta(1,\alpha)*(1-\sum_1^nw_j)$
  - This is to say that you have $beta_i \in \{\beta_1 ... \beta_n\} \sim beta(1,\alpha)$ but each successive $w_i$ is calculated as $beta_i$ multiplied by a successively shorter remaining stick lengths so that later weights tend to be smaller.
* Larger values of $\alpha$ result in weight being distributed over a larger set of smaller categories whereas as $\alpha$ appr.oaches 0, the expected value of $w_1$ approaches 1.

This process (up to a certain limit imposed by numerical precision) reserves probability mass for unobserved categories and the fact that the initial prior weights for the categories observed first is higher than that for later categories reflects the intuition that more frequent categories are most likely to be observed earlier. Using this to determine if each new observation comes from an observed or new category, we end up with:
 
* Each observation comes from a previously unobserved category with probability equal to the current stick length:
  $$p(new) \sim bernoulli(\text{stick length})$$
  - This means that the first observation comes from an unobserved category with probability 1
* If the observation came from a new category, break off a piece of the stick with length $w$ with $w = beta(1,a)*length_{current}$
  - This $w$ is the probability mass assigned initially to the new category.
* This then repeats for each new observation.

At the moment, the dispersion parameter is being setas part of the model, although it would be possible (and maybe even feasable) to have a prior distribution on this as well.

## Doing updates for category probabilities:

The conjugate prior distribution for a multinomial is a Dirichlet distribution (this is a multivariate generalization of the Beta distribution). This distribution has a vector-valued parameter $\alpha$. The components $\alpha_i$ of this vector represent counts over observations and the expected value of a Dirichlet distribution is $\left[\frac{\alpha_1}{\sum_1^n\alpha_i}...\frac{\alpha_n}{\sum_1^n\alpha_i}\right]$ corresponding to the expected value of the parameter vector $\theta$ for the multinomial distribution over category probabilities. 

Updating the dirichlet distribution after each observation is simply a matter of adding to the counts $\alpha_i$ based on the category of newly observed data.

# Algorithm:



For each observation $x$:

* The observation comes from a new category with probability equal to the current stick length
* If the observation came from an unobserved category:
  - initialize a new category with $\mu$ equal to $x$ and $\Sigma$ equal to the weighted mean of the covariance matrices of all categories.
  - de-normalize the vector of mixing weights $\phi$, extend the vector with a weight $w_{new}$ generated by breaking the stick, then re-normalize.
  - Extend the vector $\alpha$ of pseudo-counts for the the Dirichlet prior with $\alpha_{new} = w_{new}*\sum_1^n\alpha_i$ where $n$ is the number of categories previously observed.
* If the observation came from a previously observed category:
  - calculate the posterior probabilities of each category given $x$ ($p(C = c_i\vert x) = \mathcal{T}_{\nu_i-d+1}(\mu_i,\Sigma_i)$ for each category $c_i \in C$ where $C$ is the set of currently observed categories, following Gelman et al. p. 73 on the posterior predictive distribution for a normal with unknown mean and variance)
  - update the component $C_i$ which maximizes $p(C_i|x)$ where $C_i\in \{C_1...C_n\}$.
    - Alternatively, update a component $C_i$ drawn from a multinomial distribution with $p(C_i) = p(C_i|x)$

```julia
function update(M::ItterMix,x)
    # determine if new or old category
    # p(new) decreases as stick-length decreases
    isnew = rand(Bernoulli(M.stick));
    
    # generate posterior predictive densities
    # currently terribly clunky
    
    if !bool(isnew)
        post_pred = {MvTDist(float(G.nu-M.d+1),G.mu,G.sigma*((G.kappa+1)/(G.kappa*(G.nu-M.d-1)))) for G in M.components};
        # Calculate and normalize cluster responsibilities
        resp = {M.phi[i]*pdf(post_pred[i],x) for i=1:length(post_pred)};
        resp = resp/sum(resp);
        for i = 1:length(M.components);
            update(M.components[i],x,resp[i]);
            M.alpha[i] += 1*resp[i];
        end
    else
        w_new = rand(Beta(1,M.a))*M.stick;
        M.stick = M.stick - w_new;
        if M.components == {}
            sigma_new = M.mean_sigma*(M.alpha[1]);
            M.components = cat(1,M.components,{GaussND(M.d,x,sigma_new,M.alpha[1],M.alpha[1]+M.d+1)});
        else
            # Calculating phi for new category:
            # un-normalize current vector phi (divide by 1-current stick length)
            phi_un = M.phi/(1-M.stick);
            phi_un = cat(1,phi_un,{w_new});
            alpha_new = (w_new/sum(phi_un))*sum(M.alpha);
            M.alpha = cat(1,M.alpha,{alpha_new});
            sigma_new = M.mean_sigma*(alpha_new);
            M.components = cat(1,M.components,{GaussND(M.d,x,sigma_new,alpha_new+M.d,alpha_new+M.d+1)});
        end
        
    end
    M.phi = {M.alpha[i]/sum(M.alpha) for i=1:length(M.alpha)};
    mean_sigma = mean({G.sigma/(G.nu-M.d-1) for G in M.components});
end

# Competitive updating
function update(M::ItterMix,x,comp::Int)
    # determine if new or old category
    # p(new) decreases as stick-length decreases
    isnew = rand(Bernoulli(M.stick));
    
    # generate posterior predictive densities
    # currently terribly clunky
    
    if !bool(isnew)
        post_pred = {MvTDist(float(G.nu-M.d+1),G.mu,G.sigma*((G.kappa+1)/(G.kappa*(G.nu-M.d-1)))) for G in M.components};
        # Calculate and normalize cluster responsibilities
        resp = {M.phi[i]*pdf(post_pred[i],x) for i=1:length(post_pred)};
        winner = indmax(resp)
        update(M.components[winner],x);
        M.alpha[winner] += 1;
    else
        w_new = rand(Beta(1,M.a))*M.stick;
        M.stick = M.stick - w_new;
        if M.components == {}
            sigma_new = M.mean_sigma*(M.alpha[1]);
            M.components = cat(1,M.components,{GaussND(M.d,x,sigma_new,M.alpha[1],M.alpha[1]+M.d+1)});
        else
            # Calculating phi for new category:
            # un-normalize current vector phi (divide by 1-current stick length)
            phi_un = M.phi/(1-M.stick);
            phi_un = cat(1,phi_un,{w_new});
            alpha_new = (w_new/sum(phi_un))*sum(M.alpha);
            M.alpha = cat(1,M.alpha,{alpha_new});
            sigma_new = M.mean_sigma*(alpha_new);
            M.components = cat(1,M.components,{GaussND(M.d,x,sigma_new,alpha_new+M.d,alpha_new+M.d+1)});
        end
        
    end
    M.phi = {M.alpha[i]/sum(M.alpha) for i=1:length(M.alpha)};
    mean_sigma = mean({G.sigma/(G.nu-M.d-1) for G in M.components});
end
```

Posterior sampling rules based on description of Gelman et al. for posterior sampling from a multivariate normal.

```julia
# Generate posterior samples for a given component
function post_sample_gauss(G,n)
    out = zeros(G.d,n)
    temp_dist = NormalInverseWishart(G.mu,G.kappa,G.sigma,G.nu)
    for i = 1:n
        m,s = rand(temp_dist)
        out[:,i] = MvNormal(m,s)
    end
    return out
end
```
