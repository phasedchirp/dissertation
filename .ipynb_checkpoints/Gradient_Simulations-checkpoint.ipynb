```julia
using Distributions;
using PyPlot;
using StatsBase;
```

```julia
include("GAmod.jl");
include("vowels2.jl");
```

```julia
var_scaling = 0.1
iy_sample = cat(1,rand(MvNormal(iy_mu,iy_sigma*var_scaling),1000),zeros(1,1000));
ei_sample = cat(1,rand(MvNormal(ei_mu,ei_sigma*var_scaling),1000),ones(1,1000));
ae_sample = cat(1,rand(MvNormal(ae_mu,ae_sigma*var_scaling),1000),ones(1,1000)*2);
oo_sample = cat(1,rand(MvNormal(oo_mu,oo_sigma*var_scaling),1000),ones(1,1000)*3);
uw_sample = cat(1,rand(MvNormal(uw_mu,uw_sigma*var_scaling),1000),ones(1,1000)*4);
ah_sample = cat(1,rand(MvNormal(ah_mu,ah_sigma*var_scaling),1000),ones(1,1000)*5);
aw_sample = cat(1,rand(MvNormal(aw_mu,aw_sigma*var_scaling),1000),ones(1,1000)*6);
eh_sample = cat(1,rand(MvNormal(eh_mu,eh_sigma*var_scaling),1000),ones(1,1000)*7);
ih_sample = cat(1,rand(MvNormal(ih_mu,ih_sigma*var_scaling),1000),ones(1,1000)*8);
oa_sample = cat(1,rand(MvNormal(oa_mu,oa_sigma*var_scaling),1000),ones(1,1000)*9);
uh_sample = cat(1,rand(MvNormal(uh_mu,uh_sigma*var_scaling),1000),ones(1,1000)*10);
```

```julia
sample_5 = cat(2,iy_sample,oa_sample,uw_sample,ah_sample,eh_sample);
ord = randperm(size(sample_5,2));
col1 = permute!(sample_5[1,:],ord);
col2 = permute!(sample_5[2,:],ord);
col3 = permute!(sample_5[3,:],ord);
col4 = permute!(sample_5[4,:],ord);
col5 = permute!(sample_5[5,:],ord);
col6 = permute!(sample_5[6,:],ord);
col7 = permute!(sample_5[7,:],ord);
vowels_5_rand = cat(1,col1,col2,col3,col4,col5,col6,col7);
scatter(vowels_5_rand[3,:],vowels_5_rand[4,:],c=vowels_5_rand[7,:])
```

```julia
vowels_5_z = cat(1,zscore(vowels_5_rand[1:6,:],2),vowels_5_rand[7,:]);
scatter(vowels_5_z[3,:],vowels_5_z[4,:],c=vowels_5_z[7,:])
```

```julia
lims_5 = mean(vowels_5_z[2:6,:],2)[:,1]
```

```julia
Mix1 = MixInit(2,1000,lims_5[3:4]);
length(Mix1.components)
```

```julia
for i=1:3500
    update(Mix1,vowels_5_z[3:4,i],0.005);
end

j = 1
for i=1:length(Mix1.components)
    if Mix1.phi[i]>0.01
        print(j,' ', Mix1.components[i].mu)
        print('\n')
        j+=1
    end
end
```
