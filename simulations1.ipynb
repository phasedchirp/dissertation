```julia
using Distributions;
using PyPlot;
include("Diss.jl");
```

```julia
# Define both distributions
mu1 = [10.0; 0; 0];
sigma1 = [3.0 0 0; 0 6 0; 0 0 4];
mu2 = ones(3)*10;
sigma2 = [4.0 1 0; 1 2 1; 0 1 4];
clust1 = MvNormal(mu1,sigma1);
clust2 = MvNormal(mu2,sigma2);
# generate samples from both
sample1 = rand(clust1,1000);
sample2 = rand(clust2,1000);
sample_all = cat(2,sample1,sample2);
ord = randperm(2000);
col1 = permute!(sample_all[1,:],ord);
col2 = permute!(sample_all[2,:],ord);
col3 = permute!(sample_all[3,:],ord);
sample_rand = cat(1,col1,col2,col3);
```

```julia
# initialize a mixture with d = 3, a = 2, stick = 1, alpha = {10}, phi = {1}, components = {},
# mean_sigma = 3-dimensional identity matrix
mixTest = ItterMix(3,2,1,{10},{1},{},eye(3));
```

Running this with probabilistic competition:

```julia
for i=1:100
    update(mixTest,sample_rand[:,i],2);
end
mixTest.phi
```

```julia
for i=101:2000
    update(mixTest,sample_rand[:,i],2);
end
mixTest.phi
```

```julia
j=1
for i=1:length(mixTest.components)
    if mixTest.phi[i]>0.05
        print(j,' ',mixTest.components[i].mu)
        print('\n')
        j+=1
    end
end
```

This sometimes results in having multiple components for a single category (here two components for each cluster). So, not great when things don't overlap.

Trying below with deterministic competition (Each item $x$ is always assigned to the category $C$ which maximizes $p(C|x)$):

```julia
mixTest2 = ItterMix(3,2,1,{10},{1},{},eye(3)*4);

for i=1:100
    update(mixTest2,sample_rand[:,i],1);
end

mixTest2.phi
```

```julia
for i=101:200
    update(mixTest2,sample_rand[:,i],1);
end

mixTest2.phi
```

### Testing simulation with data based on Hillenbrand et al. 1995

This is just a quick test with Hertz-scaled data generated based on the parameters of the data from Hillenbrand et al. 1995. Annoyingly the values derived from the data don't quite agree with what's listed in the paper, but they're pretty close.

See vowels.ipynb for more detailed simulations with ERB-scaled data.

```julia
# parameters are in an external file because it's annoying otherwise. See Vowels.jl for details.
include("Vowels.jl")
# scaling to reduce data variability slightly because otherwise this works not so much.
var_scaling = 0.3
iy_sample = cat(1,rand(MvNormal(iy_mu,iy_sigma*var_scaling),500),zeros(1,500));
ei_sample = cat(1,rand(MvNormal(ei_mu,ei_sigma*var_scaling),500),ones(1,500));
ae_sample = cat(1,rand(MvNormal(ae_mu,ae_sigma*var_scaling),500),ones(1,500)*2);
oo_sample = cat(1,rand(MvNormal(oo_mu,oo_sigma*var_scaling),500),ones(1,500)*3);
uw_sample = cat(1,rand(MvNormal(uw_mu,uw_sigma*var_scaling),500),ones(1,500)*4);
ah_sample = cat(1,rand(MvNormal(ah_mu,ah_sigma*var_scaling),500),ones(1,500)*5);
aw_sample = cat(1,rand(MvNormal(aw_mu,aw_sigma*var_scaling),500),ones(1,500)*6);
eh_sample = cat(1,rand(MvNormal(eh_mu,eh_sigma*var_scaling),500),ones(1,500)*7);
ih_sample = cat(1,rand(MvNormal(ih_mu,ih_sigma*var_scaling),500),ones(1,500)*8);
oa_sample = cat(1,rand(MvNormal(oa_mu,oa_sigma*var_scaling),500),ones(1,500)*9);
uh_sample = cat(1,rand(MvNormal(uh_mu,uh_sigma*var_scaling),500),ones(1,500)*10);

sample_all = cat(2,iy_sample,ei_sample,ae_sample,oo_sample,uw_sample,ah_sample,aw_sample,
eh_sample,ih_sample,oa_sample,uh_sample);
ord = randperm(size(sample_all,2));
col1 = permute!(sample_all[1,:],ord);
col2 = permute!(sample_all[2,:],ord);
col3 = permute!(sample_all[3,:],ord);
col4 = permute!(sample_all[4,:],ord);
col5 = permute!(sample_all[5,:],ord);
col6 = permute!(sample_all[6,:],ord);
col7 = permute!(sample_all[7,:],ord);
vowels_rand = cat(1,col1,col2,col3,col4,col5,col6,col7);
```

Testing a posterior sampling function:

```julia
testvar = GaussND(5,[0;0;0;0;0],eye(5),0,0)
for i=1:500
    update(testvar,iy_sample[2:6,i]);
end
```

```julia
testvar.mu
```

```julia
getsigma(testvar)*3
```

```julia
function sample(G::GaussND,s_type="posterior")
    if s_type == "posterior"
        sigma = rand(InverseWishart(G.nu,G.sigma));
        mu = rand(MvNormal(G.mu,(G.sigma/G.kappa)));
        return rand(MvNormal(mu,sigma))
    else
        return rand(MvNormal(G.mu,getsigma(G)))
    end
end
```

```julia
out = Array(Float64,6,500)
for i=1:500
    out[:,i] = vcat(sample(testvar),[1]);
end
```

```julia
out2 = Array(Float64,6,500)
for i=1:500
    out2[:,i] = vcat(sample(testvar,"map"),[2]);
end

out3 = cat(2,out,out2)
```

```julia
scatter(out3[1,:],out3[2,:],c=out3[6,:])
```
